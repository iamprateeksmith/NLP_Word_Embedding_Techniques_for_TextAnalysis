### Word Embedding Techniques using Embedding Layer in Keras

```python
### Libraries USed Tensorflow> 2.0  and keras
```

```python
##tensorflow >2.0
# one_hot = one hot representation
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
sent
```

```python
### Vocabulary size
voc_size=10000
```

#### One Hot Representation

```python
# One hot Represenation -- index from dictionary
onehot_repr=[one_hot(words,voc_size)for words in sent] 
print(onehot_repr)
```

### Word Embedding Represntation

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_length=7
#embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
embedded_docs=pad_sequences(onehot_repr,padding='post',maxlen=sent_length)
print(embedded_docs)
```

```python
# Adding max no of features should be their 
dim=10
```

```python
model=Sequential()
model.add(Embedding(voc_size,10,input_length=sent_length))
# adam optimizer & Mean Squared Error
model.compile('adam','mse')
```

```python
model.summary()
```

```python
#embedded_docs[-1::]
embedded_docs[1::3]
```

```python
# Low dimensional (Embedding Matrix)
# Embedding Matrix
# voc_size = 10000, dimensions = 10 | 2000 ---->> dimensions of 10 vectors
print(model.predict(embedded_docs))
```
